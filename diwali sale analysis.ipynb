```python
# ######  EXPLORATORY DATA ANALYSIS (E D A)

# ####    GENDER
```

```python
df.columns
```

```python
ax= sns.countplot(x = 'Gender' , hue= 'Gender', data=df,
                  palette={'M': 'blue', 'F': 'red'})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', hue='Gender', y= 'Amount',data = sales_gen ,
            palette={'M': 'blue', 'F': 'red'})
```

```python
#  FROM ABOVE GRAPHS WE CAN SEE THAT MOST OF THE BUYERS ARE FEMALE AND EVEN THE PURCHASING POWER OF FEMALES ARE GREATER THAN MEN
```

```python
ax= sns.countplot(data= df,x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# #########  STATS 
```

```python
#  total numbers of orders of top 10 stats 
sales_state= df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, hue= 'State', x='State', y='Orders')
```

```python
#  total numbers of amounts/sales  of top 10 stats 
sales_state= df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, hue= 'State', x='State', y='Amount')
```

```python
#  from above graphs we can see that most of the orders & the total sales/amount are from uttar pardesh, Maharashtra and karnataka respectively
```

```python
# ### MARITAL sTATUS 
```

```python
ax= sns.countplot(data=df, hue= 'Marital_Status' ,x= 'Marital_Status')
sns.set(rc={'figure.figsize':(5,4)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state= df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data = sales_state, hue= 'Gender', x='Marital_Status', y='Amount')
```

```python
 # *from above graphs we can see that most of the buyers are married (women) and they have high purchasing power*
```

```python
# ########  OCCUPATION ########## 
```

```python
ax= sns.countplot(data=df, hue='Occupation', x='Occupation')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, hue= 'Occupation', x='Occupation', y= 'Amount')
```
